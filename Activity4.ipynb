⃣<h1>QUESTION 15
<h3>Which month has the highest total number of casualties, and how does severity vary?

```python
accidents.groupby(['month', 'Accident_Severity'])['Number_of_Casualties'].sum()
```

<h1>INSIGHT 17</h1>
<H3>Casualties are highest in May–October, with slight injuries being the most common, while serious injuries peak mid-year and fatalities are highest in August and May, highlighting the need for targeted road safety measures during these months.</H3></H3>

```python
monthly_severity = accidents.groupby(['month', 'Accident_Severity'])['Number_of_Casualties'].sum()
monthly_severity = monthly_severity.reset_index()

month_names = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June',
               7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}
monthly_severity['month'] = monthly_severity['month'].map(month_names)


monthly_severity_pivot = monthly_severity.pivot(index='month', columns='Accident_Severity', values='Number_of_Casualties')
monthly_severity_pivot.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.xlabel('Month')
plt.ylabel('Total Number of Casualties')
plt.title('Total Number of Casualties by Month and Accident Severity')
plt.legend(title='Accident Severity')
plt.tight_layout()
plt.show()
```

<h1>QUESTION 17</h1>

<H3>The number of casualties has a weak positive correlation (0.23) with the number of vehicles involved, suggesting that accidents with more vehicles tend to result in more casualties, but the impact is not strong.
