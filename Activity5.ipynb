```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('datasets\\Internet-Demographic.csv')
```

```python
df.info()
```

```python
df['Country Name']=df['Country Name'].astype('category')
df['Income Group']=df['Income Group'].astype('category')
```

```python
df.drop(columns=['Country Code'], inplace = True)
```

```python
df.dtypes
```

<h1>Question 1</h1>
<h2>Distribution of Birthrate</h2>

```python
plt.hist(df['Birth rate'], bins=10, color='pink', edgecolor='black')
plt.title('Histogram of Birth Rate')
plt.xlabel('Birth Rate')
plt.ylabel('Frequency')

plt.show()
```

<h1>Question 2</h1>
<h2>Top 5 Country with internet usage</h2>

```python
top5 = df[df['Internet users'] > 94]
top5.sort_values(by='Internet users',ascending=False)
```

<h1>Question 3</h1>
<h2>Count of Countries by Income Group</h2>

```python
income_group_counts = df['Income Group'].value_counts()

income_group_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Count of Countries by Income Group')
plt.xlabel('Income Group')
plt.ylabel('Number of Countries')
plt.xticks(rotation=45)

plt.show()
```

<h1>Question 4</h1>
<h2>Box plot of Internet Usage</h2>

```python
plt.boxplot(df['Internet users'], vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Boxplot of Internet Users')
plt.xlabel('Internet Users (%)')

plt.show()
```

<h1>Question 5</h1>
<h2>Countries with no Internet access.</h2>

```python
nointernet = df[df['Internet users'] == 20]
nointernet.sort_values(by='Internet users',ascending=False)
```

<h1>Question 6</h1>
<h2>Average Birth rate</h2>

```python
average_birth_rate = df['Birth rate'].mean()

print(f"The average (mean) birth rate is: {average_birth_rate}")
```

<h1>Question 7</h1>
<h2>Birth rate vs Internet usage</h2>

```python
plt.scatter(df['Internet users'], df['Birth rate'], color='lightblue', edgecolor='black')
plt.title('Scatter Plot: Birth Rate vs Internet Users')
plt.xlabel('Internet Users (%)')
plt.ylabel('Birth Rate')

plt.show()
```

<h1>Question 8</h1>
<h2>Countries with birth rates above 40</h2>

```python
high_birth_rate_countries = df[df['Birth rate'] > 40]
high_birth_rate_countries.sort_values(by='Birth rate',ascending=False)
```

<h1>Question 9</h1>
<h2>Pie chart of Income Groups</h2>

```python
income_group_counts = df['Income Group'].value_counts()
plt.pie(income_group_counts, labels=income_group_counts.index, autopct='%1.1f%%', colors=['red', 'blue', 'green','orange'])
plt.title('Distribution of Income Groups')
plt.show()
```

<h1>Question 10</h1>
<h2>Correlation between Birth rate and Internet Usage</h2>

```python
correlation = df['Birth rate'].corr(df['Internet users'])
correlation
print(f"The correlation between Birth rate and Internet users is: {correlation}")
```
